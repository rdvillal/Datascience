## Importar Datos desde Archivo CSV hacia SQL


En algunas ocasiones es necesario cargar los datos hacia un DBMS como PostreSQL, MySQL, Oracle, etc. buscando que los procesos de preparaci칩n, carga y actualizaci칩n de los datos sean eficientes. Utilizando la librer칤a pandas y sqlalchemy es posible realizar esta tarea

```python
import pandas as pd
from sqlalchemy import create_engine

df = pd.read_csv('C:/HOME/DATA ANALYSIS/tweet_activity_metrics_rdvillal_20211118_20211216_es.csv')
df.head()
```

Visualizamos los datos que previamente exportamos desde Twitter. Estos se leen desde un archivo separado por comas y ahora por medio de la librer칤a pandas han sido instanciados en un objeto DataFrame. Estando en el DataFrame ser치 muy f치cil cargarlo al DBMS sin necesidad de dise침ar la tabla con alg칰n comando de CREATE TABLE, los tipos de datos y estructura son autom치ticamente interpretados posteriormente con la ayuda del objeto Engine de SQLAlchemy.

Previo a la ejecuci칩n del siguiente c칩digo he instalado la base de datos postgres en localhost, en el puerto 5432, usuario postgres, contrase침a root. He creado una base de datos llamada dataengineering, ah칤 subir칠 el archivo en una nueva tabla que se llamar치 Tweets. Utilizamos index = False para evitar que tome la columna de 칤ndice que Pandas genera autom치ticamente en los DataFrame

```python
engine = create_engine('postgresql://postgres:root@localhost:5432/dataengineering')
df.to_sql('Tweets', con = engine, index = False)
```

Si se consulta en la base postgres se confirmar치 que se tiene una nueva tabla llamada Tweets ya cargada con la informaci칩n del archivo CSV. Sin embargo, ya que la instrucci칩n anterior sobreescribe la tabla en caso de que ya exista, a continuaci칩n agregar칠 otro archivo con informaci칩n con el objetivo de mostrar c칩mo ingestar datos con el mismo m칠todo sin perder los ya cargados

```python
df2 = pd.read_csv('C:/HOME/DATA ANALYSIS/tweet_activity_metrics_rdvillal_20220801_20220901_es.csv')
df2.head()
```
